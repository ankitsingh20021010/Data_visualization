#pie chart
it is a type of chart use to visualize proportions of whole.
It is best use for showing Percentage distribution

```python
import matplotlib.pyplot as plt
```

```python
items = ["Pizza", "Burger", "Pasta"]
sales = [45, 30, 20]
```

```python
plt.pie(sales, labels=items, autopct='%1.1f%%')
plt.title("Food Sales")
plt.show()
```
